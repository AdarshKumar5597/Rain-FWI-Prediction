## Data Cleaning

```python
dataset.isnull().sum()
```

```python
## missing values
dataset[dataset.isnull().any(axis = 1)] # --> returns the column where null values are present
```

The data set is converted into two sets based on Region, from 122nd index we can make a new column based on the Region
1. 'Bejajia Region Dataset'
2. 'Sidi-Bel Abbes Region Dataset'
3. Add new column with Region

```python
dataset.loc[0:122, ['Region']] = 0 # 0 for 'Bejajia Region Dataset' region
dataset.loc[122: , ['Region']] = 1 # 1 for 'Sidi-Bel Abbes Region Dataset' region
df = dataset
```

```python
df.head()
```

```python
df['Region'] = df['Region'].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop = True) # dropping the null values as only 2 rows are with null values so it won't affect model much
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# remove the 122nd row
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip(' ')
```

```python
list(df.columns)
```

```python
df.info()
```

```python
## changing the data-types of required columns to integer datatype
columns = [column for column in df.columns if df[column].dtype == 'O']
```

```python
columns
```

```python
df.head()
```

```python
# separating columns - which columns to be converted to float or int
columns_to_int = columns[:columns.index('Ws')+1]
columns_to_float = columns[columns.index('Ws')+1: len(columns)-1]
columns_to_int, columns_to_float
```

```python
# changing the datatype of the column
df[columns_to_int] = df[columns_to_int].astype(int)
df[columns_to_float] = df[columns_to_float].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
# save the dataframe to csv file to make a copy for model training
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index = False)
```
